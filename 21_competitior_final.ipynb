```python
print('Do you want to play a game? Please type Y or N')
answer = input().upper()
if answer == 'Y' :
    print("Okay, let's play!")
    
    print('choose a random number N, (make this number at least as big as 20 for a fun game)')
    N = input()
    
    print('choose a random number x which is less than N, (for an interesting game make this number larger than 2).')
    x = input()
        
    print('We are going to play a game where we take turns counting up to '+ str(N) + 
      '. We can each say up to ' + str(x) +' numbers in one go, for example if you start on your go you could count up to ' 
      + str(1) + ' or ' + str(x) + ' or any number in between. Whoever says ' + N+ ' looses.') 
    
    total_moves = N
    m = int(x) + 1
    y = (int(N) - 1)%int(m)
    temp = 0
    counter = 0
    
    if y == 0 :
        print("Alright, you're up first!")
        player = 2
    else:
        print("Alright, I'll go first!")
        player = 1
        
        
        
    if player == 1:
        print("Okay I count up to the number " + str(y) + ".")
        counter = int(y)
        turn_next = False
    else:
        print("What number do you count up to?")
        temp = int(input())
        
        while (int(temp) - int(counter) > int(x)) or (int(temp) <= int(counter)):
            print("Uh oh, that's an invalid move. Please try again.")
            temp = 0
            print("What number do you count up to?")
            temp = int(input())
            
        counter = int(temp)
        adds = int(counter)
        turn_next = True
        
        
        
    while int(counter) < int(total_moves):
        if turn_next == True:
            counter = int(counter) + int(m) - int(adds)
            print('Okay my turn. I count up to the number ' + str(counter) + ' . Your turn.')
            turn_next = False
        else:
            print("What number do you count up to?")
            temp = int(input())
            
            while (int(temp) - int(counter) > int(x)) or (int(temp) <= int(counter)):
                print("Uh oh, that's an invalid move. Please try again.")
                temp = 0
                print("What number do you count up to?")
                temp = int(input())
            
            
            up_to = temp
            adds = int(up_to) - int(counter)
            counter = int(up_to)
            turn_next = True

            
            
    print('I win! Better, luck next time ;)!')
        
        
else:
    print('Okay, play you some other time then.')
```
